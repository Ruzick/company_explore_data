```python
def get_rev_num(url,comp):
    options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(executable_path=r'C:/Users/kikik/capstone/chromedriver.exe', options=options)
    driver.get(url)
    num_reviews = driver.find_element_by_xpath('.//div[@class="row "]').text
    #\W+\w+(.*?)\w+
#     a = re.search(r'Overview\W+(.*?)\w+',num_reviews).group(1)
    a = re.search(r'Overview\W+(.*?\W+.*?\w+)',num_reviews).group(1)
    a= re.sub('\n', '', a)
    a=re.sub('Reviews','',a)
#     reviews[comp]=a
    return (comp,a)
```
